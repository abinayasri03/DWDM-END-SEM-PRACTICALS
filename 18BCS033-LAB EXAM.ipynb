```python
#creating the training and target variables

x=data.drop("output",axis=1)
y=data["output"]
```

```python
#Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
#Fitting classifier(model) to the Training set¶

from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
#Predicting the Test set results

prediction=model.predict(x_test)
prediction
```

```python
prediction=model.predict([[68,1,0,144,193,1,1,141,0,3.4,1,2,3]])
print("predicted value :", str(prediction))
#result:not suffered from heart disease
```

```python
prediction=model.predict([[57,0,0,120,354,0,1,163,1,0.6,2,0,2]])
print("predicted value :", str(prediction))
#result:suffered from heart disease
```

```python
#Confusion matrix

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,prediction)
cm
```

```python
#Predicting the model accuracy

from sklearn import metrics
print("accuracy",metrics.accuracy_score(y_test,prediction))
```

```python
test_prediction=model.predict(x_test)
metrics.accuracy_score(test_prediction,prediction)
```

```python

print(metrics.classification_report(y_test,test_prediction))
```
